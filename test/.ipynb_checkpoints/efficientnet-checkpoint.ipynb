```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.utils import Sequence, to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Activation, Add
from tensorflow.keras.layers import BatchNormalization, Dropout, GlobalAveragePooling2D, GlobalMaxPooling2D
from tensorflow.keras.layers import ZeroPadding2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.applications import EfficientNetB0, EfficientNetB1, EfficientNetB2
from tensorflow.keras.datasets import cifar10
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import albumentations as A
import sklearn
from sklearn.model_selection import train_test_split
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
class CIFAR10_Dataset(Sequence):
    
    def scaling_and_resize(self, img, height, width):
        img = (img / 255.).astype("float32")
        return cv2.resize(img, (width, height))
    
    def __init__(self, image_list, label_list, height, width, batch_size, shuffle):
        self.image_list = image_list
        self.label_list = label_list
        self.height = height
        self.width = width
        self.batch_size = batch_size
        self.shuffle = shuffle
        if self.shuffle:
            self.on_epoch_end()
    
    def __len__(self):
        return int(np.ceil(len(self.image_list) / self.batch_size))
    
    def __getitem__(self, idx):
        images = self.image_list[idx * self.batch_size:(idx + 1) * self.batch_size]
        image_batch_list = []
        label_batch = self.label_list[idx * self.batch_size:(idx + 1) * self.batch_size]
        
        for img in images:
            img = self.scaling_and_resize(img, height=self.height, width=self.width)
            image_batch_list.append(img)
        
        image_batch = np.array(image_batch_list)
        
        return image_batch, label_batch

    def on_epoch_end(self):
        if self.shuffle:
            idxs = np.arange(len(self.image_list))
            np.random.shuffle(idxs)
            self.image_list = self.image_list[idxs]
            self.label_list = self.label_list[idxs]
```

```python
image_train, image_valid, label_train, label_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=2023)
```

```python
train_ds = CIFAR10_Dataset(image_train, label_train, height=224, width=224, batch_size=32, shuffle=True)
valid_ds = CIFAR10_Dataset(image_valid, label_valid, height=224, width=224, batch_size=32, shuffle=False)
```

```python
efficientb0_model = EfficientNetB0(include_top=True, weights=None, input_shape=(224, 224, 3), classes=10)
efficientb0_model.summary()
```

```python
efficientb0_model.compile(optimizer=Adam(learning_rate=0.00001), loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
rlr_cb = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, mode='min', verbose=1)
ely_cb = EarlyStopping(monitor='val_loss', patience=10, mode='min', verbose=1)
mdcp_cb = ModelCheckpoint(filepath="./weights/effb0_epoch_{epoch:02d}_val_loss_{val_loss:.3f}.h5",\
                         save_weights_only=True, monitor="val_loss", mode="min")
```

```python
history = efficientb0_model.fit(train_ds, epochs=20, validation_data=valid_ds, callbacks=([rlr_cb, ely_cb, mdcp_cb]), verbose=1)
```
