## preprocess

```python
def preprocess_fn(img, augment=None, height=224, width=224, normalize=True):
    
    def pad_to_square(img):
        h, w, c = img.shape
        if h == w:
            return img
        if h < w:
            pad1, pad2 = (w - h + 1) // 2, (w - h) // 2
            return cv2.copyMakeBorder(img, pad1, pad2, 0, 0, cv2.BORDER_WRAP)
        pad1, pad2 = (h - w + 1) // 2, (h - w) // 2
        return cv2.copyMakeBorder(img, 0, 0, pad1, pad2, cv2.BORDER_WRAP)
    
    img = pad_to_square(img)
    
    if augment is not None: # test_ds: augment=None
        img = augment(img)
        
    img = cv2.resize(img, (width, height))
    
    if normalize:
        img = (img / 255.).astype("float32")
    
    return img
```

```python
def augment_fn(img):
    transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=20, border_mode=cv2.BORDER_WRAP, p=0.5),
        A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
        A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5)
    ])
    
    return transform(image=img)["image"]
```

## dataset on `tf.keras.utils.Sequence`

```python
class Flower_Dataset(Sequence):
    
    def __init__(self, imgpath_list, label_list, height=224, width=224,\
                 batch_size=64, preprocess=None, augment=None, shuffle=True):
        self.imgpath_list = imgpath_list
        self.label_list = label_list
        self.height = height
        self.width = width
        self.batch_size = batch_size
        self.preprocess = preprocess
        self.augment = augment
        self.shuffle = shuffle
        if self.shuffle:
            self.on_epoch_end()
    
    def __len__(self):
        return int(np.ceil(len(self.imgpath_list) / self.batch_size))
    
    def __getitem__(self, idx):
        imgpath_batch = self.imgpath_list[idx * self.batch_size:(idx + 1) * self.batch_size]
        image_batch_list = []
        label_batch = self.label_list[idx * self.batch_size:(idx + 1) * self.batch_size]
        
        for imgpath in imgpath_batch:
            img = cv2.cvtColor(cv2.imread(imgpath), cv2.COLOR_BGR2RGB)
            if self.preprocess is not None: # test_ds: augment=None
                img = preprocess(img=img, augment=self.augment, height=self.height, width=self.width, normalize=True)
            
            image_batch_list.append(img)
        
        image_batch = np.array(image_batch_list)
        
        return image_batch, label_batch

    def on_epoch_end(self):
        if self.shuffle:
            idxs = np.arange(len(self.imgpath_list))
            np.random.shuffle(idxs)
            self.imgpath_list = self.imgpath_list[idxs]
            self.label_list = self.label_list[idxs]
```

```python
def train_valid_dataset_split(data_df, height=224, width=224, batch_size=64,\
                        preprocess=None, augment=None):
    train_df, valid_df = train_test_split(data_df, test_size=0.2, random_state=2023, stratify=data_df["label"])
    imgpath_train, one_hot_enc_train = train_df["imgpath"].values, train_df.iloc[:, 2:].values
    imgpath_valid, one_hot_enc_valid = valid_df["imgpath"].values, valid_df.iloc[:, 2:].values
    
    train_ds = Flower_Dataset(
        imgpath_list=imgpath_train, label_list=one_hot_enc_train,
        height=height, width=width,
        batch_size=batch_size,
        preprocess=preprocess,
        augment=augment,
        shuffle=True
    )
    
    valid_ds = Flower_Dataset(
        imgpath_list=imgpath_valid, label_list=one_hot_enc_valid,
        height=height, width=width,
        batch_size=batch_size,
        preprocess=preprocess,
        augment=None,
        shuffle=False
    )
    
    return train_ds, valid_ds
```

```python
train_ds, valid_ds = train_valid_dataset_split(data_df, height=224, width=224,\
                                              batch_size=64, preprocess=preprocess_fn, augment=augment_fn)
```
